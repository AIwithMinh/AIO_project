```python
TRAINED_MODEL_PATH = '/content/yolov10/yolov10n.pt'
model = YOLOv10(TRAINED_MODEL_PATH)
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 256
model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```

```python
from google.colab.patches import cv2_imshow

IMAGE_URL = '/content/drive/MyDrive/71f4fa9fe4214d99ae38b1412eb63a53.jpg'
CONF_THRESHOLD = 0.5
IMG_SIZE = 640
results = model.predict(source=IMAGE_URL,
                       imgsz=IMG_SIZE,
                       conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
```
